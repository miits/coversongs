```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open('raw_scores.array', 'rb')
scores = pickle.load(f)
f.close()
```

```python
true = scores.diagonal()
true.shape
```

```python
false = np.array([])
for i in range(1, 80):
    false = np.append(false, scores.diagonal(-i))
    false = np.append(false, scores.diagonal(i))
false.shape
```

```python
x = np.append(true, false)
y = np.append(np.ones(shape=(80,)), np.zeros(shape=(6320,)))
x = x.reshape((-1, 1))
y = y.reshape((-1, 1))
data = np.append(x, y, axis=1)
```

```python
f = open('raw_scores_labeled.dataset', 'wb')
pickle.dump(data, f)
f.close()
```

```python
l_true = np.linspace(1, len(true), num=len(true))
plt.plot(true, l_true, 'bo')
plt.show()
```

```python
l_false = np.linspace(1, len(false), num=len(false))
plt.plot(false, l_false, 'ro')
plt.show()
```
